## World Cup Data Analysis

```
# File location and type
file_location = "/FileStore/tables/WorldCups__1_.csv"
file_type = "csv"

df = spark.read.csv(file_location,header=True,inferSchema=True)
df.show()
```

```
df.printSchema()
```

```
df.columns
```

```
#country with most goals

sort_df = df.sort(df['GoalsScored'].desc())
sort_df[['Country','GoalsScored']].show(1)
```

```
#country that have won the worldcup the most 

country_count = df.groupBy('Winner').count()
country_most = country_count.sort(country_count['count'].desc())
country_most.show(1)
```

```
#most runnerup
country_count = df.groupBy('Runners-Up').count()
country_run = country_count.sort(country_count['count'].desc())
country_run.show(3)
```

```
#The total goal scored in the world cup

print("total goals scored is:{}".format(df.groupBy().sum().collect()[0][1]))
```

```
#convert attendance to integer
df = df.withColumn("Attendance",df.Attendance.cast('int'))
```

```
#total attendance
print("total attendance is:{}".format(df.groupBy().sum().collect()[0][4]))
```

```
#check sum of all numerical columns
df.groupBy().sum().show()
```

```
#where year is 1930
df.where(df['Year']=="1930").show()
```
